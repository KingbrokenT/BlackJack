```python
import random
```

```python
#We are just going to take the identifiers for cards from the Warmup excercises. No need to reinvent the wheel
suits = ('Hearts', 'Diamonds', 'Spades', 'Clubs')
ranks = ('Two', 'Three', 'Four', 'Five', 'Six', 'Seven', 'Eight', 'Nine', 'Ten', 'Jack', 'Queen', 'King', 'Ace')
values = {'Two':2, 'Three':3, 'Four':4, 'Five':5, 'Six':6, 'Seven':7, 'Eight':8, 
            'Nine':9, 'Ten':10, 'Jack':10, 'Queen':10, 'King':10, 'Ace':11}
```

```python
#The most important thing about our card objects is their value, but keeping a suit and rank will help us properly
#generate a deck.
class Card:
    
    def __init__(self,suit,rank):
        self.suit = suit
        self.rank = rank
        self.value = values[rank]
    
    def revalue(self,new_val):
        self.value = new_val
        
    def __str__(self):
        return self.rank + ' of ' + self.suit
```

```python
#Decks need to generate themselves and hold cards. Our deck will be set up so that it empties out as the games go on
class Deck:
    
    def __init__(self):
        # Note this only happens once upon creation of a new Deck
        self.all_cards = [] 
        for suit in suits:
            for rank in ranks:
                # This assumes the Card class has already been defined!
                self.all_cards.append(Card(suit,rank))
                
    def shuffle(self):
        # Note this doesn't return anything
        random.shuffle(self.all_cards)
        
    def deal_one(self):
        # Note we remove one card from the list of all_cards
        return self.all_cards.pop()    
```

```python
#We need two players, a dealer and a player
class Player:
    
    def __init__(self,name):
        self.name = name
        # A new player has no cards
        self.all_cards = []
        self.size = len(self.all_cards)
        
    def discard(self):
        # Note we remove one card from the list of all_cards
        # We state 0 to remove from the "top" of the deck
        # We'#ll imagine index -1 as the bottom of the deck
        return self.all_cards.clear()
    
    def add_cards(self,new_cards):
        #We will use this command to deal and hit
        if type(new_cards) == type([]):
            self.all_cards.extend(new_cards)
        else:
            self.all_cards.append(new_cards)
    
    
    def __str__(self):
        return f'{self.name} has {len(self.all_cards)} cards.'
```

```python
You = Player('You')
dealer = Player('Dealer')
bjdeck = Deck()
bjdeck.shuffle()
game = True
gameon = True
chips = 1000
print('Welcome to blackjack. You have 1000 Chips.')

#First ask for a bet and reduce chips by the bet amount
#Deal starting hand to dealer and player. Reveal the player's cards and 1 dealer card
#ask Player hit or stay. 
    #If hit deal 1 card, check for a bust
    #If bust, player loses, if not, ask again
#Dealer will hit until the net value is 17 (and on a soft 17)
    #check each hit for a bust
#Ace will be an 11 until it would bust the hand
def cont(chips):
    while True:
        ans = input(f'You have {chips} chips remaining. Would you like to play another round, Y or N?: ')
        if ans == 'Y':
            game =  True
            return game
        elif ans == 'N':
            print('Thank you for playing')
            game = False
            return game
        else:
            print('Invalid Response')
You.discard()
dealer.discard()
while game == True:
    #Take a bet, check that it is an interger
    is_int = False
    while not is_int:
        bet_pr = input('Enter your bet: ')
        try:
            float(bet_pr)
            if float(bet_pr) == int(bet_pr):
                bet = int(bet_pr)
                if bet <= chips:
                    is_int = True
                else:
                    print(f'Please enter a bet that is less than your remaining chips({chips})')
                    is_int = False
            else:
                print('Please enter an interger bet.')
                is_int = False
        except:
            print('Please enter an interger bet.')
            is_int = False


    #Deal a new had to each palyer and return the player, dealer and bet
    for card in range(2):
        You.add_cards(bjdeck.deal_one())
        dealer.add_cards(bjdeck.deal_one())
    print(f'You have: {You.all_cards[0]} and {You.all_cards[1]}. The dealer has {dealer.all_cards[0]}')

    #The player hits
    hit = True
    while hit:
        pcard_values = []
        for card in You.all_cards:
            pcard_values.append(card.value)
            pcard_sum = sum(pcard_values)
            if pcard_sum > 21 and 11 in pcard_values:
                card.revalue(1)
                pcard_values[pcard_values.index(11)] = 1
                pcard_sum = sum(pcard_values)
            elif pcard_sum > 21:
                print(f'You drew {card}. Your total is now {pcard_sum}. You have bust and lost')
                chips -= bet
                game = cont(chips)
                You.discard()
                dealer.discard()
                hit = False
                gameon = False
                break
        if gameon == True:
            while True:
                hit_ans = input(f'Your total is {pcard_sum}. Would you like to hit, Y or N?: ')
                if hit_ans.upper() == 'Y':
                    hit = True
                    You.add_cards(bjdeck.deal_one())
                    break
                elif hit_ans == 'N':
                    print('You stay')
                    hit = False
                    break
                else:
                    print('Invalid Response')

    if gameon == True:
        #The Dealer hits
        hit = True
        while hit:
            dcard_values = []
            for card in dealer.all_cards:
                dcard_values.append(card.value)
                dcard_sum = sum(dcard_values)
                if dcard_sum > 21 and 11 in dcard_values:
                    card.revalue(1)
                    dcard_values[dcard_values.index(11)] = 1
                elif dcard_sum > 21:
                    print(f'The Dealer drew {card}. Their total is now {dcard_sum}. They have bust and you have won!')
                    chips += bet
                    game = cont(chips)
                    You.discard()
                    dealer.discard()
                    hit = False
                    gameon = False
                    break
            if gameon == True:
                if dcard_sum <= 17:
                    hit = True
                    dealer.add_cards(bjdeck.deal_one())
                    print(f'The Dealer drew {card}.')
                else:
                    print('The Dealer stays')
                    hit = False
            
    if gameon == True:
        #The Reveal
        print(f'The Dealer has {dcard_sum} and you have {pcard_sum}.')
        if pcard_sum > dcard_sum:
            print('You have won!')
            chips += bet
            game = cont(chips)
            You.discard()
            dealer.discard()
        elif pcard_sum < dcard_sum:
            print('You have lost!')
            chips -= bet
            game = cont(chips)
            You.discard()
            dealer.discard()
        else:
            print('It was a tie!')
            game = cont(chips)
            You.discard()
            dealer.discard()
```
